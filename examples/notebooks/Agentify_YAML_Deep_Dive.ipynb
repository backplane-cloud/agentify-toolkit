This notebook focuses entirely on **YAML-driven agents**, the core of Agentify.

## Philosophy

Agentify treats agents as **configuration artifacts**.

- Human-readable
- Version-controlled
- Runnable without Python code


## Full YAML Schema Example


````
agent_yaml = '''
name: security-reviewer
description: Security Architecture Reviewer
version: 0.1.0
model:
  provider: openai
  id: gpt-5-nano
  api_key_env: OPENAI_API_KEY
role: |
  You are a senior AI security architect.
  Review designs for security risks and mitigations.
'''

print(agent_yaml)
````

## Running YAML Agents

```bash
agentify run agent.yaml
```

Or run all agents in a folder:

```bash
agentify run agents/
```

## Loading YAML into Python


````
from agentify import Agent

agent = Agent.from_yaml_string(agent_yaml)
agent.run("Review a zero trust network design")
````

## Best Practices

- One agent per YAML file
- Use clear, role-based prompts
- Treat YAML as production config
- Avoid embedding secrets
